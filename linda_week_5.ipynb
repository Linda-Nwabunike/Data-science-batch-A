```python
#load libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url="https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
df=pd.read_csv(url)
print(df)
```

```python
data=df
plt.scatter(x=data.sepal_length, y=data.sepal_width)
plt.show()
```

```python
#scattered plot using seaborn
sns.relplot(x="petal_length",y="petal_width",hue="species",data=df)
```

```python
#using matplotlib to plot histogram of three features
plt.hist(x=df['species'])
plt.title=('Histogram distribution of species')
plt.show()
```

```python
# using seaborn and matplotlib to plot a histogram of two features
sns.histplot( x=df ["species"], y=df ["petal_length"])
plt.xlabel('species')
plt.ylabel('petal_length')
plt.title=('histogram distribution of petal_length')
plt.show()
```

```python
sns.histplot(data=df,x="sepal_length",y="sepal_width", color="red",bins=25)
```

```python
sns.lineplot(data=df,x="petal_length",y="petal_width")
plt.show()
```
